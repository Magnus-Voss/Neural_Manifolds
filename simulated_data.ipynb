```python
import os
import scipy.io as io
import numpy as np
import matplotlib.pyplot as plt
codeDir = os.getcwd()
len2del = len(codeDir.split('/')[-1])
homeDir = codeDir[:-len2del]
file_name = homeDir+'dataset/data.mat'
mat = io.loadmat(file_name)
```

```python
timeseries = list(mat['data'][0][0])[0] # Calcium Time series N x T
position = list(mat['data'][0][0])[1][0] # Animal position (degrees)
tvals = list(mat['data'][0][0])[2][0] # time points (10Hz)
```

```python
#visualise animal position
plt.figure(figsize=(20, 2))      
plt.plot(position)
plt.axis('off')

#Visualise all cell traces as final check
from itertools import cycle
cycol = cycle('bgrcmk')
columns = 1
rows = 200
fig, ax_array = plt.subplots(rows, columns, figsize=(20, 10), squeeze=False)
for i,ax_row in enumerate(ax_array):
    for j,axes in enumerate(ax_row):
        axes.set_facecolor('xkcd:white')
        axes.plot(timeseries[i].astype(float), c='k', linewidth=1)
        axes.get_yaxis().set_ticks([])
        axes.get_xaxis().set_visible(False)
        axes.axis('off')
```
